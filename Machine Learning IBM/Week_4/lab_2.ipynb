## ***Example $(1)$.***

In this first example, we will show how to prove (or disprove), with statistical evidence, that the BMI of females is different from that of males.

$ \ $

### $(a)$ ***Choose a sample statistic***


The first step in hypothesis testing is to choose a sample test statistic. Hypothesis testing allows us to check the sample statistic against a statistic of another sample or population. Let  $\mu_{1}$ and $\mu_{2}$ given by:

* $\color{aquamarine}{\mu_{1}}$ = "the population mean for BMI of males"

* $\color{aquamarine}{\mu_{2}}$ =  "the the population mean for BMI of females"

We will compare these mean values statistically.


$ \ $

### $(b)$  ***Define hypothesis (Null and Alternative)***

The next step is to define the hypothesis to be tested. Hypothesis is defined in two ways:

$ \ $

$(\checkmark)$ ***Null hypothesis***

Null hypothesis is a statistical hypothesis which assumes that the difference in observations is due to a random factor. It is denoted by $\color{aquamarine}{H_{0}}$.

$ \ $


$(\checkmark)$ ***Alternative hypothesis***

Alternative hypothesis is the opposite of null hypothesis. It assumes that the difference in observations is the result of a real effect. The alternate hypothesis is denoted by $\color{aquamarine}{H_{1}}$.

$ \ $

In this cases, we will take as hypothesis:


* $\color{aquamarine}{ùêª_{0}=\mu_{1}-\mu_{2} = 0}$ which means that  "There is no difference between the BMI of male and BMI of female."

* $\color{aquamarine}{ùêª_{A} =\mu_{1}-\mu_{2} \neq 0}$ which means that  "There is difference between the BMI of male and BMI of female."

$ \ $

## ***Statistics to use in this lab***

In this lab, we will use one of the $t$-test, $z$-score, $f$-score or chi-squared statistics to evaluate our results.

$ \ $

$(\checkmark)$ A $t$-test is used for testing the mean of one population against a standard or comparing the means of two populations if you do not know standard deviation of the the population and when you have a limited sample ($n < 30$). If you know the standard deviation of the populations , you may use a $z$-test.

$ \ $

$(\checkmark)$ A $z$-test is used for testing the mean of a population versus a standard, or comparing the means of two populations, with large ($n ‚â• 30$) samples, whether you know the population standard deviation or not. It is also used for testing the proportion of some characteristic versus a standard proportion, or comparing the proportions of two populations.

$ \ $

$(\checkmark)$ An $f$-test is used to compare variances between 2 populations. The samples can be any size. It is the basis of ANOVA.

$ \ $

$(\checkmark)$ chi-squared test is used to determine whether there is a statistically significant difference between the expected and the observed frequencies in one or more categories of a contingency table. A contingency table is a tabular representation of categorical data. It shows the frequency distribution of the variables.

## ***Null hypothesis and alternative hyphotesis***

Now, let's define our null and alternative hypothesis.

$ \ $


$(\checkmark)$ $\color{aquamarine}{ùêª_{0}:\mu_{1} \leq \mu_{2}}$  $\Leftrightarrow$ The average charges of smokers are less than or equal to nonsmokers. 

$ \ $

$(\checkmark)$ $\color{aquamarine}{ùêª_{A}:\mu_{1} > \mu_{2}}$  $\Leftrightarrow$ The average charges of smokers are greater than or equal to nonsmokers.

$ \ $

The '>' sign in the alternate hypothesis indicates the test is right tailed. To compare the mean values of smoking and nonsmoking populations, we will use a $t$-test. If $z$-values (calculated from a t-test) fall into the area on the right side of a distribution curve, this would cause us to reject the null hypothesis.

$ \ $

$(1)$ Now, let's plot our smoking versus nonsmoking populations by using *seaborn* `boxplot()` function. It is always useful to have a visual representation of the data that we are working with.


```python
data.head()
```

```python
sns.boxplot(x = data["charges"], y = data["smoker"], data = data).set(title="Fig:1 Smoker vs Charges")
```

$ \ $

$(2)$ Now, we will calculate $t$-value and $p$-value of charges for smoking and nonsmoking populations.


```python
alpha = 0.05


# smoker_char = smoker["charges"]
# nonsmoker_char = nonsmoker["charges"]
t_val, p_value = stats.ttest_ind(smoker_char, nonsmoker_char)


# we divided by 2 due the previous imagen
p_value_onetail = p_value/2


# we show on the screen the results
print(f"t_value = {t_val} , p_value ={p_value} , p_value_onetail = {p_value_onetail}")
```

$ \ $


$(3)$ Use `print()` function to state your conclusions based on the calculated statistics.


```python
if p_value <alpha : 
  print(f"Conclusion: Since p_value {p_value_onetail} is less than alpha {alpha}")
  print("\n")
  print("Reject null hypothesis that average charges for smokers are less than or equal to nonsmokers.")
else:
   print(f"Conclusion:Since p value {p_value_onetail} is greater than alpha {alpha}") 
   print("\n")
   print("Failed to reject null hypothesis that average charges for smokers are less than nonsmoker.")
```

Now, let's define our null and alternative hypothesis.

$(\checkmark)$ $\color{aquamarine}{ùêª_{0}:\mu_{1}=\mu_{2}=\mu_{3}}$  $\Leftrightarrow$ "The mean BMI of women with no children , one child, and two children are the same."

$(4)$ Now, we will construct the ANOVA table and check for each groups count (0,1,2 children) against the bmi values. We will use the ols (ordinary least squares) model for estimation of the unknown parameters.


* To learn more about OLS (ordinare lest squares), please visit this wikipedia [link](https://en.wikipedia.org/wiki/Ordinary_least_squares?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMML0232ENSkillsNetwork30654641-2022-01-01).


```python
formula = 'bmi ~ C(children)'


model = ols(formula, female_children).fit()


aov_table = anova_lm(model)


aov_table
```

$(\checkmark)$ we fail to reject the null hypothesis (no rechazamos la hip√≥tesis nula $H_{0}$).

$(\checkmark)$ the mean bmi of women with no children, one child, and two children are the same.

## ***Example*** $(4)$

In this example, we will determine if the proportion of smokers is significantly different across the different regions.

$ \ $


## ***Null hypothesis and alternative hyphotesis***

Now, let's define our null and alternative hypothesis.

$ \ $

$(\checkmark)$ $\color{aquamarine}{ùêª_{0}}$  $\Leftrightarrow$ "Smokers proportions are not significantly different across the different regions." 

$ \ $

$(\checkmark)$ $\color{aquamarine}{ùêª_{A}}$  $\Leftrightarrow$ "Smokers proportions are different across the different regions."

$ \ $

Here, we are comparing two different categorical variables, smokers/nonsmokers and different regions. For this type of analysis, we will perform a chi-square test.

$ \ $

$(1)$ First, we will calculate a contingency table between the proportions of smokers in different regions. For this, we will use *pandas* `crosstab()` function.






```python
contingency= pd.crosstab(data["region"], data["smoker"])
contingency
```

$ \ $

$(2)$ Next, let's plot the distribution of nonsmokers/smokers across 4 different regions using the `plot()` function.


```python
contingency.plot(kind='bar')
```

```python
contingencia = contingency.reset_index()
contingencia.index.rename('Index', inplace=True)
contingencia
```

$ \ $

$(3)$ Now, using `chi2_contingency()` method, from the *scipy.stats* `chi2_contingency`, we will calculate chi-squared, p-value, degrees of freedom, and expected frequencies for our data.


```python
chi2, p_val, dof, exp_freq = chi2_contingency(contingency, correction = False)

print(f"chi-square statistic: {chi2}")
print("\n")

print(f"p_value: {p_val}")
print("\n")

print(f"expected frequencies: \n {exp_freq}")
```

$ \ $

$(4)$ Based on the above results, print your conclusion statements whether to reject or accept the null hypothesis. 


```python
if p_val < 0.05: 
  print('Reject the null hypothesis, that the smokers proportions are not significantly different across the different regions')   
else: 
  print('Accept the null hypothesis, that the smokers proportions are not significantly different across the different regions')
```

$ \ $

$(5)$ What are your conclusions about the hypothesis?

$ \ $

We failed to reject that the proportions of smokers are not significantly different across different regions. Therefore, the proportions of smokers are different across different regions.
