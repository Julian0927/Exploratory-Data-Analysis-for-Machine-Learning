## ***Conducting a formal significance test for one of the hypotheses and discuss the results***





```python
tips_data
```

```python
df_male = tips_data[df["sex"]=="Male"]
df_female = tips_data[df["sex"]=="Female"]
```

```python
total_male = df_male["total_bill"]
total_female = df_female["total_bill"]
```

```python
sns.distplot(total_male, color = 'green', hist = False)
sns.distplot(total_female, color = 'red', hist = False)
```

$ \ $


## ***Hypothesis***


$(1)$ Let  $\mu_{1}$ and $\mu_{2}$ given by:

* $\color{aquamarine}{\mu_{1}}$ = "the mean for the total_bill of males"

* $\color{aquamarine}{\mu_{2}}$ =  "the mean for the total_bill of females"

We will compare these mean values statistically.


$ \ $

$(2)$ In this cases, we will take as hypothesis:


* $\color{aquamarine}{ð»_{0}=\mu_{1}-\mu_{2} = 0}$ which means that  "There is no difference between the total_bill given by males and total_bill given by  females."

* $\color{aquamarine}{ð»_{A} =\mu_{1}-\mu_{2} \neq 0}$ which means that  "There is difference between the total_bill given by males and total_bill given by  females."

```python
df_male["total_bill"].mean()
```

```python
df_female["total_bill"].mean()
```

```python
alpha = 0.05

t_value, p_value = stats.ttest_ind(total_male,  total_female)

print("t_value = " , t_value)
print("p_value = " , p_value)
```

```python
if p_value < alpha:
    print(f"Conclusion: since p_value {p_value} is less than alpha {alpha} \n")

    print("We reject the null hypothesis that there is no difference between total_bill given by males and the total_bill given by females.")
    
else:
    print(f"Conclusion: since p_value {p_value} is greater than alpha {alpha} \n")
    print("We fail to reject the null hypothesis that there is  difference between total_bill given by males and the total_bill given by females")
```
