```python
best_query = """SELECT playerID, sum(GP) as "num_games_played", AVG(startingPos) as "avg_starting_position"
                FROM allstarfull
                GROUP BY playerID
                ORDER BY num_games_played DESC, avg_starting_position ASC
                LIMIT 3"""

best = pd.read_sql(best_query, con)
print(best.head())
```
